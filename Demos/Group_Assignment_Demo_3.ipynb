```python
import pickle as pkl
import os 

import pandas as pd
import numpy as np

import datetime

import re

import matplotlib.pyplot as plt
import seaborn as sns
```

# 1. Import Data

```python
# import the data
hvc_hvvisitresult = pd.read_csv("../final_data/HVC_HVVISITRESULT.csv", sep=";")
hvc_so0 = pd.read_csv("../final_data/HVC_SO0.csv", sep=";")
hvc_am0 = pd.read_csv("../final_data/HVC_AM0.csv", sep=";")
hvc_ar0 = pd.read_csv("../final_data/HVC_AR0.csv", sep=";")
hvc_hvvisitresultdetails = pd.read_csv("../final_data/HVC_VISITRESULTDETAILS.csv", sep=";")
hvc_hvposition = pd.read_csv("../final_data/HVC_HVPOSITION.csv", sep=";")
```

```python
# convert date
hvc_hvvisitresult["DATEONLY"] = pd.to_datetime(hvc_hvvisitresult["DATEONLY"])
```

# 2. Explorotary Analytics

## 2.1. Transaction amounts

First, lets inspect the mean, std, median, first quartile, third quartile and range of the transaction amount

```python
# get the transactions with a succesfull outcome
pos_transactions = hvc_hvvisitresult[hvc_hvvisitresult["HVOUTCOME_NRID"] == 2]["AMOUNT"]
```

```python
# mean 
print(np.mean(pos_transactions))

# standard deviation 
print(np.std(pos_transactions))

# median transaction
print(np.median(pos_transactions))

# first quartile
print(np.percentile(pos_transactions, 25))

# third quartile
print(np.percentile(pos_transactions, 75))

# range
print(np.ptp(pos_transactions))
```

Now, let's plot a boxplot of the transaction amounts in order to explore the median, 1st and 3th quartile and outliers

```python
# plot boxplot of amount
sns.boxplot(pos_transactions)
plt.show()
```

Lets also make a boxplot for the total sales generated by the customers for each season type

```python
# join transactions with customers
cust_transactions = pd.merge(hvc_hvvisitresult, hvc_so0, on="SO0_NRID", how="inner")

# get transactions with succesfull outcome
cust_transactions = cust_transactions[cust_transactions["HVOUTCOME_NRID"] == 2]

# create boxplot for transaction amounts for each season type
sns.boxplot(x=cust_transactions["AMOUNT"], y=cust_transactions["SEASON_TYPE"])
plt.show()
```

Now lets also have a look at the evolution of the transactions amounts during the year 2019

```python
# get all the transactions for 2019
transactions_2019 = hvc_hvvisitresult[(hvc_hvvisitresult["DATEONLY"] >= pd.Timestamp(year=2019, day=1, month=1)) & (hvc_hvvisitresult["DATEONLY"] < pd.Timestamp(year=2020, month=1, day=1))]

# get all the transactions with succesfull outcome
transactions_2019 = transactions_2019[transactions_2019["HVOUTCOME_NRID"] == 2]

# get total sales per day
total_sales_per_day = transactions_2019.groupby("DATEONLY")["AMOUNT"].sum().reset_index()

# sort the total sales per day
total_sales_per_day = total_sales_per_day.sort_values(by="DATEONLY")

# get the number of the day
total_sales_per_day["day_idx"] = total_sales_per_day["DATEONLY"].apply(lambda x: (x - pd.Timestamp(year=2019, day=1, month=1)).days)

# create dataframe with all days of the year
df_all_days = pd.DataFrame({"day_idx": range(365)})

# join datasets to get sales for every day in 2019
joined = pd.merge(left=df_all_days, right=total_sales_per_day, on="day_idx", how="left")
# replace missings with 0
joined = joined.fillna({"AMOUNT": 0})


# plot time series

# initialize figure
plt.figure(figsize=(8, 5))
# define title
plt.title("Total Sales Per Day In 2019")
# plot the transaction amount per day
plt.plot([obs if obs != 0 else None for obs in joined["AMOUNT"]])
# add labels
plt.ylabel("Total Sales")
plt.xlabel("Day")
# add xticks
plt.xticks(range(0, 365, 30), range(0, 365, 30), rotation=45)
plt.show()
```

Now lets plot a barplot of the total sales generated per month in 2020

```python
# get the month of each transaction date and store it in new column
hvc_hvvisitresult["MONTH"] = hvc_hvvisitresult["DATEONLY"].apply(lambda x: x.month)

# get transactions from 2020 only
transactions_2020 = hvc_hvvisitresult[(hvc_hvvisitresult["DATEONLY"] >= pd.Timestamp(year=2020, day=1, month=1)) & (hvc_hvvisitresult["DATEONLY"] < pd.Timestamp(year=2021, month=1, day=1))]

# get transactions with succesfull outcome
transactions_2020 = transactions_2020[transactions_2020["HVOUTCOME_NRID"] == 2]

# get the total sales for each month
monthly_sales = transactions_2020.groupby("MONTH")["AMOUNT"].sum().reset_index()

# plot barplot

# define figure canvas
plt.figure(figsize=(8, 5))
# plot barplot
plt.bar(x=range(12), height=monthly_sales["AMOUNT"])
# set xticks as the first three letters of the name of each month
month_names = ["JAN", "FEB", "MAR", "APR", "MAY", "JUN", "JUL", "AUG", "SEP", "OCT", "NOV", "DEC"]
plt.xticks(range(12), month_names)
# define title
plt.title("Monthly Sales of 2020")
plt.show()
```

## 2.2. Customer Positions

Now that we have explored the customer transactions, lets investigate the positions of the customers

```python
# check positions
hvc_hvposition.head(5)
```

```python
# plot coordinates on scatter plot
plt.figure(figsize=(8, 8))
plt.scatter(hvc_hvposition["LONG"], hvc_hvposition["LAT"], s=6)
plt.show()
```

Next, we will use KMeans clustering to try to cluster these customers into different regions

```python
# import kmeans model
from sklearn.cluster import KMeans

# initialize kmeans model with 3 clusters
km_model = KMeans(n_clusters=3)

# fit model on data
km_model.fit(hvc_hvposition[["LONG", "LAT"]])

# get cluster labels for each customer
cluster_labels = km_model.predict(hvc_hvposition[["LONG", "LAT"]])

# check first 5 cluster labels
cluster_labels[:5]
```

Lets now plot the clusters by using different colors for each cluster

```python
# define the colors for each cluster
color_dict = {0: "red", 1: "blue", 2: "green"}

# get the cluster color for each customer
cluster_colors = [color_dict[label] for label in cluster_labels]

# plot the customer locations with the corresponding colors
plt.figure(figsize=(8, 8))
plt.scatter(hvc_hvposition["LONG"], hvc_hvposition["LAT"], color=cluster_colors, s=6)
plt.show()
```

## 2.3. Routes

Next, lets have a quick look at the routes. More specifically, we are going to visualize which customers are on a specific route

```python
# define a route id
route_id = 219020601.0

# join customer info with customer positions
cust_pos = pd.merge(hvc_so0, hvc_hvposition, on="SO0_NRID", how="inner")

# get all the customers belonging to that route
all_cust_route = cust_pos[cust_pos["HVROUTETEMPLATE_NRID"] == 219020601.0]

# plot the positions of all customers 
plt.figure(figsize=(8, 8))
plt.scatter(hvc_hvposition["LONG"], hvc_hvposition["LAT"], color="blue", s=6)
# plot the positions of the customers belonging to the specified route
plt.scatter(all_cust_route["LONG"], all_cust_route["LAT"], color="orange", label="customer of route %s" %route_id, s=7)
# define legend to add label to figure
plt.legend(loc="lower left")
plt.show()
```
